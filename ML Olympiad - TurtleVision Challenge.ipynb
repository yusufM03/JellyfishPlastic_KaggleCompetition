```python
import numpy as np
import cv2
import random

# Function to apply random rotation
def random_rotation(image):
    angle = random.randint(-15, 15)
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

# Function to apply random horizontal flip
def random_horizontal_flip(image):
    if random.random() < 0.5:
        image = cv2.flip(image, 1)
    return image

# Function to apply random vertical flip
def random_vertical_flip(image):
    if random.random() < 0.5:
        image = cv2.flip(image, 0)
    return image


# Load and preprocess your images
augmented_images = []
for i,image in enumerate(X):
    augmented_image = random_horizontal_flip(image)
    a= random_vertical_flip(augmented_image )
    
    
    augmented_images.append(augmented_image)
    augmented_images.Append(a)
    y.append(y[i])
    y.append(y[i])
    

# Convert list of augmented images to numpy array
augmented_images = np.array(augmented_images)
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV, train_test_split
import numpy as np

# Define your Keras model as a function
def create_model(optimizer='adam', dropout_rate=0.2, units=64):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (4, 4), activation='relu', input_shape=(244, 244, 3)))
    model.add(layers.Conv2D(120, (3, 3), activation='relu', padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (2, 2), activation='relu', kernel_regularizer=regularizers.l2(0.01)))
    model.add(layers.BatchNormalization())
    model.add(layers.GlobalAveragePooling2D())
    model.add(layers.Dense(units, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dense(6, activation='softmax'))
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Split data into training and validation sets


# Define hyperparameters and their search space
param_grid = {
    'optimizer': ['adam', 'rmsprop'],
    'dropout_rate': [0.1, 0.2, 0.3],
    'units': [64, 128, 256]
}

# Create Keras classifier
model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32, verbose=1)

# Perform random search with validation data
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=10, cv=3, verbose=1)
random_search_result = random_search.fit(X_train, y_train_encoded, validation_data=(X_val, y_val))

# Print results
print("Best: %f using %s" % (random_search_result.best_score_, random_search_result.best_params_))
```
